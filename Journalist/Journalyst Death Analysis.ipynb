<h2>📰 The Risks Journalists Face</h2>

<p>There are certain areas in the world that pose a <span class='highlight'>high risk</span> for journalists. These places can be dangerous for both <span class='highlight'>foreign</span> and <span class='highlight'>local</span> reporters, though the level of risk may vary.</p>

<h3>🎯 The Dangers of Journalism</h3>
<p>Journalism comes with <span class='highlight'>inherent risks</span>, especially when covering <span class='highlight'>sensitive topics</span>. The <span class='highlight'>medium of reporting</span>, the <span class='highlight'>nature of the coverage</span>, and even the journalist's <span class='highlight'>nationality</span> can all influence the level of danger they face.</p>

<h3>🎥 The Struggles of Freelance Journalists</h3>
<p>Freelance journalists, in particular, may be at greater risk. Without the <span class='highlight'>institutional support</span> and <span class='highlight'>resources</span> available to full-time employees, they might find themselves in more <span class='highlight'>precarious situations</span>, lacking both <span class='highlight'>protection</span> and <span class='highlight'>legal backing</span>.</p>

<h3>❌ Impunity and Suppression</h3>
<p>Many journalists have been <span class='highlight'>punished, kidnapped, or subjected to violence</span> for reporting on <span class='highlight'>controversial subjects</span>. In some cases, the perpetrators of these attacks enjoy <span class='highlight'>complete impunity</span>.</p>

#### 2.2 Coverage Per Sub Region

```python
def classify_job (job):
    if 'reporter' in job.lower():
        return 'Reporter'
    elif 'producer' in job.lower():
        return 'Producer'
    elif 'publisher' in job.lower():
        return 'Publisher'
    elif 'editor' in job.lower():
        return 'Editor'   
    elif 'photographer' in job.lower():
        return 'Photographer'
    elif 'camera operator' in job.lower():
        return 'Camera operator'
    elif 'technician' in job.lower():
        return 'Technician'
    elif 'columnist' in job.lower():
        return 'columnist'
    else:
        return 'Unknown'


job_class = []
for i,job in data[['jobs']].iterrows():
    job_class.append(classify_job(job[0]))
```

```python
data['job_class'] = job_class
data['job_class'].unique()
```
